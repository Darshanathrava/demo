```python
# linear algebra
from numpy import linalg as lg
import numpy as np

a = np.array([[5, 4],
              [1, 2]])

print("array is:\n", a)

# Calculate eigenvalues and eigenvectors of the real symmetric matrix
eigenvalues, eigenvectors = lg.eig(a)

print("eigenvalues are:\n", eigenvalues)
print("eigenvectors are:\n", eigenvectors)
```

```python
#decision tree
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
df = pandas.read_csv("D_tree.csv")
d = {'uk': 0, 'usa': 1, 'na': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'y': 1, 'n': 0}
df['Go'] = df['Go'].map(d)
features = ['Age', 'Experence', 'Rank', 'Nationality']
X = df[features]
Y = df['Go']
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, Y)
tree.plot_tree(dtree, feature_names=features)
plt.show()
```
